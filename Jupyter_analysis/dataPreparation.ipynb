# Data collection 


Source : (https://www.kaggle.com/datasets/joebeachcapital/30000-spotify-songs/data)

Playlists having an appropriate number of tracks are being selected for the project. At first, it was decided to collect the playlist with number of of tracks between 20 to 30 because I am doing the data splitting later in the evaluation and for each playlist, if the number of tracks is split into 80%-20%, that would give 16 to 24 tracks for input and 4 to 6 tracks for comparison which I think is an appropriate amount. However, there was only a few playlist (which contains between 20-30 tracks) and therefore, the total tracks were only 436 which is quite small. 
Therefore, I adjust the number of tracks in a playlist to be in a range 15-40 and that would be 12 to 32 tracks for input and 3 to 8 tracks for comparison and the total tracks is 1315 before data cleaning. 

```python
import pandas as pd
playlistData = pd.read_csv ("spotify_songs.csv", encoding="latin1")

# Select only a part of the playlists

# playlist_size table => playlist ids and number of tracks per playlist (count)
playlist_size = playlistData.groupby(playlistData['playlist_id']).size().reset_index(name='count')

# target_playlist_size table => playlist ids with size between 15 and 40 tracks
target_playlist_size = playlist_size [(playlist_size['count'] > 15) & (playlist_size['count'] < 40)]

# Cumulative count of tracks in target playlists
target_playlist_size['cumulative_count'] = target_playlist_size['count'].cumsum()

target_playlist = playlistData[playlistData['playlist_id'].isin(target_playlist_size['playlist_id'])]

target_playlist

```

## Tracks file creation

Select fields for Tracks file. 
Audio features, track id, name and artist name will be included in the Tracks file.


```python
tracks = target_playlist[['track_id', 'track_name', 'danceability', 'energy', 'valence', 'acousticness', 'tempo', 'instrumentalness','loudness']]
tracks.head()
```

## Artist Name and ID fetching from Spotify API

The original dataset uses '&' to store two or more artists for a track. Even if we split the two artists using delimiter '&', there can still be artists/bands with '&' in their name. So, I decided not to use the artist name (track_artist) provided by this dataset as it is not suitable to split the data using '&'.

Therefore, artist name and id are fetched through spotify track id. Both of them are stored as lists to handle multiple artists involving in a track.

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id="YOUR_CLIENT_ID",
    client_secret="YOUR_CLIENT_SECRET"
))

# fetching artist data function : give track id as input, to get artist id and name as an output
def get_artist_data(track_id):
    try: 
        track = sp.track(track_id)
        artist_id = [artist['id'] for artist in track['artists']]       #get artist ids list
        artist_name = [artist['name'] for artist in track['artists']]       #get artist names list
    except:
        artist_id = None
        artist_name = None
    return artist_id, artist_name

# store to the tracks dataframe
tracks[['artist_id', 'artist_name']] = tracks['track_id'].apply(lambda t: pd.Series(get_artist_data(t)))
tracks
```

## Data Preprocessing 
1. data cleaning : no nan values are found, no cleaning needed.
2. some changes to track name
2. getting to know the data

```python
import re

# check if any nan values is there
print ("total nan values in df: ",tracks.isna().sum())

regexPatternsList = [r'\s*-\s*From.*\s*["\'].*["\']',            # - From ..." ... "   
                r'\s*-\s*Theme Song From\s*["\'].*["\']' ,       # - Theme Song From "..."
                r'\s*-\s*Main Title Theme From\s*["\'].*["\']', # - Main Title Theme From "..."
                r'\s*\(\s*Theme from .*\)\s*.*',                     # (Theme from ...) 
                r'\s*[\[\\(]\s*feat\..*[\]\\)]',                                  # (feat. ...) or [feat. ...]
                r'\s*-\s*feat\..*',                                      # feat.
                r'\s*\(\s*with.*\)',                                     # (with ...)
                r'\s*\(\s*.*Remix.*\)',                                     # (... Remix ...)
                r'\s*-.*Remix.*',                                         # - ... Remix ...
                r'\s*-\s*[Remastered|Remasterizado].*',                 # - Remastered ... or - Remasterizado...
                r'\s*\(\s*Remastered\s*.*\)',            # ( Remastered ....) 
                r'\s*-\s*\d{4}\s*Remaster(ed)?\s*',  # - 2008 Remaster 
                r'\s*\(\s*Original Mix\s*\)',          # (Original Mix)
                r'\s*-\s*Original Mix.*'              # - Original Mix
]
regexPatterns = '|' . join(regexPatternsList)

tracks['fixed_track_name'] = tracks['track_name'].str.replace(regexPatterns, '', flags=re.IGNORECASE, regex=True)

tracks.describe()
```

## Features to Vector
1. Normalization
2. Weighting

## 1. Normalization
min-max normalization is used. Sklearn has MinMaxScalar() and it is used to normalize the features. 



```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# features to be normalized
features_vector = ['danceability', 'energy', 'valence', 'acousticness', 'tempo', 'instrumentalness', 'loudness']
print("Original  :", tracks.iloc[0][features_vector].tolist())
normalized = scaler.fit_transform(tracks[features_vector])      # 2D array
tracks["normalized_vector"] = normalized.tolist()       # transform to list to add to dataframe
tracks.head()
print("Normalized  :", tracks.iloc[0]['normalized_vector'], type(tracks.iloc[0]['normalized_vector']))

```

## 2. Weighting

Energy, valence and acousticness can better discriminate the songs. (1.4 and 1.3 weightings are given to those.)

Instrumentalness and loudness are better than danceability and tempo at discriminating songs. (1.2 and 1.1 weightings are given respectively.)

```python
weight_list = np.array([
        1.1,  # danceability
        1.4,  # energy
        1.3,  # valence     
        1.3,  # acousticness
        1.1,  # tempo
        1.2,  # instrumentalness
        1.2   # loudness
])

# Weighting : multiply each normalized vector by its corresponding weight (numpyarray * numpyarray) -> convert back to python list
tracks["weighted_vector"] = tracks["normalized_vector"].apply(lambda norm_v: (np.array(norm_v) * weight_list).tolist())

tracks.head()

print("Original  :", tracks.iloc[0][features_vector].tolist())
print("Normalized:", tracks.iloc[0]["normalized_vector"])
print("Weighted  :", tracks.iloc[0]["weighted_vector"])
```
